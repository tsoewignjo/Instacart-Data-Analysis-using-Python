#### Step 1: Perform Data Cleaning

Steps for data cleaning:
    1. Fix column names
    2. Check missing values
    3. Check for duplicates
    4. Check for mixed data type
    

```python
# Data Wrangling: Fix column names
customers.rename(columns = {'First Name' : 'first_name'}, inplace = True)
customers.rename(columns = {'Surnam' : 'last_name'}, inplace = True)
customers.rename(columns = {'Gender' : 'gender'}, inplace = True)
customers.rename(columns = {'STATE' : 'state'}, inplace = True)
customers.rename(columns = {'Age' : 'age'}, inplace = True)
customers.rename(columns = {'date_joined' : 'date_join'}, inplace = True)
customers.rename(columns = {'n_dependants' : 'dependents'}, inplace = True)
customers.rename(columns = {'fam_status' : 'married'}, inplace = True)
```

```python
customers.head()
```

```python
# Show null values in first name column
customers[customers['first_name'].isnull()]
```

**Note:** Since the missing of these first names will not affect our analysis, we will do nothing about this.

```python
# Check for duplicates
customers[customers.duplicated()]
```

**Note:** No duplicate

```python
# Check for mixed-type data in your df_ords dataframe.

for col in customers.columns.tolist():     
    # variable weird is to be used to determine there is mixed data type
    weird = (customers[[col]].map(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (customers[weird]) > 0:
        print (col)
```

**Note:** There is a mixed data type in first_name column. This maybe because of the Nan values.

#### Step 2: Combine Customer data with prepared InstaCart data

```python
df_combined.shape
```

```python
# Merging Instacart data with Customer data using user_id as the merging key 
print ('Start time: ' + datetime.now().strftime("%H:%M:%S"))
df_merged = df_combined.merge(customers, on = 'user_id')
print ('End time: ' + datetime.now().strftime("%H:%M:%S"))
```
